```python
import kagglehub

# ä¸‹è½½ IMDB æ•°æ®é›†
path = kagglehub.dataset_download("lakshmi25npathi/imdb-dataset-of-50k-movie-reviews")

print("æ•°æ®é›†å·²ä¸‹è½½ï¼Œæœ¬åœ°è·¯å¾„ä¸ºï¼š", path)
```

```python
import pandas as pd
import os

# è‡ªåŠ¨æ‰¾åˆ°æ•°æ®é›†æ–‡ä»¶
csv_path = os.path.join(path, "IMDB Dataset.csv")
df = pd.read_csv(csv_path)
df.head()
```

```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', str(text))  # å»é™¤HTMLæ ‡ç­¾
    text = re.sub(r'[^a-zA-Z0-9.,!?\s]', '', text)  # å»é™¤ç‰¹æ®Šå­—ç¬¦
    text = text.lower().strip()  # è½¬å°å†™ã€å»ç©ºæ ¼
    return text

df["cleaned_review"] = df["review"].apply(clean_text)
print("æ¸…æ´—å®Œæˆï¼")
df.head()
```

```python
# å»æ‰é‡å¤å’Œç©ºè¡Œ
df = df.drop_duplicates(subset="cleaned_review").dropna(subset=["cleaned_review"])

# ä¿å­˜åˆ° data æ–‡ä»¶å¤¹
output_path = "../data/imdb_cleaned.csv"
os.makedirs("../data", exist_ok=True)
df.to_csv(output_path, index=False)
print(f"å·²ä¿å­˜æ¸…æ´—ç»“æœåˆ°: {output_path}")
print("æœ€ç»ˆæ ·æœ¬æ•°é‡:", len(df))
```
