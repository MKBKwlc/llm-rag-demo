```python
import kagglehub

# 下载 IMDB 数据集
path = kagglehub.dataset_download("lakshmi25npathi/imdb-dataset-of-50k-movie-reviews")

print("数据集已下载，本地路径为：", path)
```

```python
import pandas as pd
import os

# 自动找到数据集文件
csv_path = os.path.join(path, "IMDB Dataset.csv")
df = pd.read_csv(csv_path)
df.head()
```

```python
import re

def clean_text(text):
    text = re.sub(r'<.*?>', '', str(text))  # 去除HTML标签
    text = re.sub(r'[^a-zA-Z0-9.,!?\s]', '', text)  # 去除特殊字符
    text = text.lower().strip()  # 转小写、去空格
    return text

df["cleaned_review"] = df["review"].apply(clean_text)
print("清洗完成！")
df.head()
```

```python
# 去掉重复和空行
df = df.drop_duplicates(subset="cleaned_review").dropna(subset=["cleaned_review"])

# 保存到 data 文件夹
output_path = "../data/imdb_cleaned.csv"
os.makedirs("../data", exist_ok=True)
df.to_csv(output_path, index=False)
print(f"已保存清洗结果到: {output_path}")
print("最终样本数量:", len(df))
```

# Day1: 数据清洗总结
- 使用 kagglehub 下载 IMDB 数据集（50k 条影评）
- 进行了 HTML 标签清理、符号过滤、小写化
- 去除重复项，保存为 data/imdb_cleaned.csv
